```python
hint_list = []
for i in contents_json:
    hint_list.append(requests.get(i[0]['url']) )
```

```python
for i in hint_list:
    print(i.json()['name'])
```

```python
hint_order = re.compile('64|00([0-9][0-9])|[1-9][1-9]')
```

```python
ordered_dict = {}
for i in hint_list:
    ordered_dict[i.json()['name']] = hint_order.findall(i.json()['name'])
```

```python
ordered_dict
```

```python
ordered_list = sorted(ordered_dict.items(), key=lambda x:x[1])
```

```python
sorted_dict = dict(ordered_list)
```

```python
sorted_dict
```

```python
for i in sorted_dict:
    for x in hint_list:
        if x.json()['name'] == i:
            print(x.json()['content'])
```

```python
import base64
for i in sorted_dict:
    for x in hint_list:
        if x.json()['name'] == i:
            print(base64.b64decode((x.json()['content'])))
```
