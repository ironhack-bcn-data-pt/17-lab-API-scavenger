### Challenge 3: Hidden Cold Joke
#### Find out the cold joke contained in the 24 secret files in the  repo: 'ironhack-datalabs/scavenger' 

#### The filenames of the secret files contain '.scavengerhunt' and are named from '.0001.scavengerhunt' to '.0024.scavengerhunt'. They are scattered randomly throughout the repo. 

```python
# We will be using the GitHub Search API that allows you to search for the specific item
```

```python
GITHUB_API_BASE_URL = 'https://api.github.com/'
GITHUB_API_SEARCH_CODE_PATH = 'search/code?q='
codeFileQuery = '.scavengerhunt+in:path+repo:ironhack-datalabs/scavenger'

query_url = (GITHUB_API_BASE_URL + GITHUB_API_SEARCH_CODE_PATH + codeFileQuery)
resp_3 = requests.get(query_url, auth=(github_user, token))
result_3 = resp_3.json()
```

```python
result_3['items'][0]
```

```python
scavengerhunt = []

for file in result_3['items']:
    scavengerhunt.append(file['url'])
    scavengerhunt = sorted(scavengerhunt)

scavengerhunt
```

```python
r = re.compile("(00\d\d)")
scavengerhunt.sort(key=lambda x: int(r.search(x).group()))
```

```python
scavengerhunt
```

```python
# Now we need to read the content of each of the scavengerhunt files 
for i in scavengerhunt:
    resp_4 = requests.get(i, auth=(github_user, token))
    result_4 = resp_4.json()
    string = result_4['content']
    print(string)   
```

```python
# Converting the the encoded content of the file from base64 to ascii 
content = []

for i in scavengerhunt:
    resp_4 = requests.get(i, auth=(github_user, token))
    result_4 = resp_4.json()
    string = result_4['content']
    string = str(base64.b64decode(string))
    string = string.replace("\\n'", "").replace("b'","")
    content.append(string)
```

```python
content
```

```python
# Concatenating the strings and printing the joke
joke = ' '.join(content)
print(joke)
```
